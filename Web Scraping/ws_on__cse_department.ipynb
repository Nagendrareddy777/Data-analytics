**Libraries**

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
link="https://rguktn.ac.in/academics/departments/cse/"
```

```python
link
```

# sending request to link


```python
raw_data=requests.get(link)
```

```python
raw_data
```

```python
raw_data.text
```

**Converting raw data to html formatted data to understand easily**

```python
soup=BeautifulSoup(raw_data.text,"html.parser")
```

```python
soup
```

**process to scrap data from given link**

```python
faculty_details=[]
for faculty in soup.find_all("div",id="faculty")[0].find_all("div",class_="col-md-6 weather-top"):
  image_link  ="https://rguktn.ac.in/academics/departments/"+faculty.find("img").get("src")[2:]

  profile_link=faculty.find("a").get("href")[2:]
  name        =faculty.find("a").text
  role        =faculty.find("span",class_="f-s-15").text
  qualification=faculty.find("span",class_="").text
  mail_id     =faculty.find_all("span",class_="f-s-15")[1].text

  faculty_details.append([name,qualification,role,profile_link,image_link,mail_id])
```

```python
faculty_details
```

**converting scrapped data to dataframe**

```python
df=pd.DataFrame(faculty_details,columns=["NAME","QUALIFICATION","ROLE","PROFILE_LINK","IMAGE_LINK","MAIL_iD"])
```

```python
df["NAME"].head()
```

**converting dataframe to excel file which stores in local system**

```python
df.to_excel("CSE_DEPARTMENT.xlsx",index=False)
```
