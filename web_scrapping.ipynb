```python
author_Data=soup.find_all("div",class_="quote")
```

```python
author_Data[1]
```

```python
author_list=[]
for i in author_Data:
  quote=i.find("span",class_="text").text[1:-1]
  author=i.find("small",class_="author").text
  author_id=i.find("a").get("href")
  tags=[]
  for tag in i.find_all("a",class_="tag"):
    tags.append(tag.text)
  author_list.append([quote,author,author_id,tags])
```

```python
entire_data=[]
for page in range(1,11):
  link="https://quotes.toscrape.com/page/"+str(page)
  res=requests.get(link)
  soup=BeautifulSoup(res.text,"html5lib")
  for quote in soup.find_all("div",class_="quote"):
    q=quote.find("span",class_="text").text[1:-1]
    author=quote.find("small",class_="author").text
    author_id=quote.find("a").get("href")
    tags=[]
    for tag in quote.find_all("a",class_="tag"):
      tags.append(tag.text)

    entire_data.append([q,author,author_id,tags])
```

```python
entire_data
```

```python
len(entire_data)
```

```python
df=pd.DataFrame(entire_data,columns=["quote","author","author_id","tags"])
```
